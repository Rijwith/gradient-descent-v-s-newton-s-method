```python
# Gradient descent method.
import sympy
lst_step=[]
lst_func=[]  # These lists are defined to create a bargraph.
lst_functions=[]
```

```python
from sympy import *
x,y=symbols('x y')
def decent(func,x_0,step_size,tolerance,colour): # INPUTS REQUIRED : FUNC, INTIAL POINT, STEP SIZE, TOLERANCE AND COLOUR OF THE GRAPH.
    diff=Derivative(func,x).doit()
    x_1=x_0-diff.subs(x,x_0)*step_size
    lst2=[x_0,x_1]
    lst1=[0,1]
    count=1
    title=str(func)+" with step size "+str(step_size)+" and tolerance "+str(tolerance)+" and initial point "+str(x_0)
    lst_functions.append(func)
    while abs(x_1-x_0)>tolerance:
        x_1,x_0=(x_1-(diff.subs(x,x_1))*step_size),x_1
        lst2.append(x_1)
        count=count+1
        lst1.append(count)
    plt.plot(lst1,lst2,marker="o",color=colour,label=title)
    plt.title(str(func))
    plt.xlabel("no of iterationS")
    plt.ylabel("value of x_n")
    plt.grid()
    plt.legend(bbox_to_anchor=(1.44,-0.25),loc='upper right')
    plt.tight_layout()
    lst_step.append(count)
    lst_func.append(title)
```

```python
from sympy import*
x,y=symbols('x y')
decent(x**2+4*x+1,4,0.1,0.1,"red")
decent(x**2+4*x+1,5,0.2,0.01,"yellow")
decent(x**2+4*x+1,6,0.3,0.001,"green")
plt.show()
decent(exp(x)-log(x),1,0.1,0.1,"yellow")
decent(exp(x)-log(x),1,0.2,0.3,"orange")
decent(exp(x)-log(x),1,0.3,0.2,"black")
```

```python
from sympy import*
x,y=symbols('x y')
def newton(func,x_0,tolerance,colour): # INPUTS REQUIRED : FUNC,INTIAL POINT, STEP SIZE, TOLERANCE AND COLOUR OF THE GRAPH.
    diff=Derivative(func,x).doit()
    x_1=x_0-func.subs(x,x_0)/diff.subs(x,x_0)
    lst2=[x_0,x_1]
    lst1=[0,1]
    count=1
    title=str(func )+" with tolerance "+str(tolerance)+" and initial point "+str(x_0)

    while abs(x_1-x_0)>tolerance:
        x_1,x_0=float(x_1-(func.subs(x,x_1))/diff.subs(x,x_1)),x_1
        lst2.append(x_1)
        count=count+1
        lst1.append(count)
    
    plt.plot(lst1,lst2,marker="o",color=colour,label=title)
    plt.title(str(func))
    plt.xlabel("no of iteration")
    plt.ylabel("value of x_n")
    plt.grid()
    plt.legend(bbox_to_anchor=(1.44,-0.25),loc='upper right')
    plt.tight_layout()
    lst_steps.append(count)
    lst_func2.append(title)
```

```python
from sympy import*
x,y=symbols('x y')
newton(x**2+4*x+1,4,0.01,"red")
newton(x**2+4*x+1,5,0.01,"yellow")
newton(x**2+4*x+1,6,0.01,"green")
plt.show()
newton(exp(x)-log(x),6,1,"green")
newton(exp(x)-log(x),4,1,"violet")
newton(exp(x)-log(x),5,1,"blue")
```
